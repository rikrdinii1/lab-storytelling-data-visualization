## Bonus Challenge 1 - Adjust Data Distribution

If you used the correct method in the previous step, you should have found the data distribution is skewed to the left. In order to improve your data visualization in the next steps, you can opt to adjust the `SalePrice` column by applying a mathematical function to the values. The goal is to produce a bell-shape normal distribution after applying the mathematical function to the sale price.

*This technique is optional in data visualization but you'll find it useful in your future machine learning analysis.*

#### In the cell below, adjust the `SalePrice` column so that the data are normally distributed.

Try applying various mathematical functions such as square root, power, and log to the `SalePrice` column. Visualize the distribution of the adjusted data until you find a function that makes the data normally distributed. **Create a new column called `SalePriceAdjusted` to store the adjusted sale price.**

[This reference](https://trainingdatascience.com/workshops/histograms-and-skewed-data/) shows you examples on how to adjust skewed data.

```python
# your code here
x_min = df['SalePrice'].min()
x_max = df['SalePrice'].max()

df['SalePriceAdjusted'] = (df['SalePrice']-x_min)/(x_max-x_min)
df['SalePriceAdjusted'].hist(bins=50)
```

Note that:

    * Visualizing how the number of half bathrooms affected the sale price probably does not make sense. You can create a new column to calculate the total number of bathrooms/rooms then visualize with the calculated number.

```python
# your code here
# add cells as needed

#Size of the house (GrLivArea, LotArea, and GarageArea)
x=df['SalePrice']

y1 = df['GrLivArea']
y2 = df['LotArea']
y3 = df['GarageArea']

fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
a = sns.scatterplot(ax=axes[0],data=df,x=x,y=y1)
axes[0].set_title(y1.name)

b = sns.scatterplot(ax=axes[1],data=df,x=x,y=y2)
axes[1].set_title(y2.name)

c = sns.scatterplot(ax=axes[2],data=df,x=x,y=y3)
axes[2].set_title(y3.name);

# Aqui podemos interpretar que no es realmente importante el tamaño salvo el del lote, todos los demás tienen una relacion
# muy lineal con los precios de venta
```

```python
#Number of rooms (BedroomAbvGr, KitchenAbvGr, FullBath, HalfBath, BsmtFullBath, BsmtHalfBath)
x=df['SalePrice']

df['CuartosTot']=df['BedroomAbvGr']+df['KitchenAbvGr']+df['FullBath']+df['HalfBath']+df['BsmtFullBath']+df['BsmtHalfBath']
y = df['CuartosTot']

sns.scatterplot(data=df,x=x,y=y,hue=y);
# Tambien podemos ver que si hay una cierta relacion entre el numero de habitaciones totales en el precio de venta de la casa.
```

```python
#How long the house has been built or remodeled (YearBuilt and YearRemodAdd)
y=df['SalePrice']
df['AñosConstruccion'] = 2021 - df['YearBuilt']
df['AñosRemodelacion']=2021 - df['YearRemodAdd']

fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=True)
a = sns.scatterplot(ax=axes[0],data=df,y=y,x=df.AñosRemodelacion)
axes[0].set_title('Años Remodelacion')

b = sns.scatterplot(ax=axes[1],data=df,y=y,x=df.AñosConstruccion)
axes[1].set_title('Años Construccion')

# En Años Remodelacion podemos ver que están topados entre 65 y 80 años de antiguedad, yaun asi la mayor parte de los datos
# se concentran en un precio bajo. 
# mientras que en Años de Construccion, podemos ver que no necesariamente una mayor antiguedad reprecenta un mayor precio, 
# al contrario, entre mas reciente sea la casa, mas valor tendrá
```

```python
# Neighborhood of the house (Neighborhood)
x=df['SalePrice']
y = df['Neighborhood']

fig, axes = plt.subplots(1, 1, figsize=(15, 10), sharey=True)
a = sns.scatterplot(data=df,y=y,x=x,hue=y)
a.set_title('Neighborhood');

# Aqui claramente podemos ver como cambia la distribucion de precios por cada uno de los vecindarios, dandonos a 
# entender, que efectivamente tenemos vecindarios donde podemos encontrar casas mas economidas, casas mas caras, o un equilibrio
```

## Bonus Challenge 2 - Exploring Data with Correlation Heatmap

Now you have explored data visualizations with certain fields based on common sense. In the dataset there are many other fields that you are not sure whether they are important factors for the sale price. What is the best way to explore those fields without investigating them individually?

Making scatter matrix is not an option here because there are too many fields which makes it extremely time consuming to create scatter matrix. One option you have is to create a heatmap. Heatmaps are much less expensive to create than scatter matrixes. You can use heatmaps to visualize the pairwise correlations between each two variables.

Here is a [reference](https://seaborn.pydata.org/examples/many_pairwise_correlations.html) you can use to learn how to creat the pairwise correlation heatmap. Your heatmap should look like below:

![Corr Heatmap](heatmap.png)

```python
# your code here
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

In your heatmap, you can easily identify the highly correlated (either positively or negatively) variables by looking for the grids with darker colors. 

#### In the cell below, summarize what variables are highly correlated to the sale price?

```python
# your comment here

sale_price_corr = corr['SalePrice']
sale_price_corr = pd.DataFrame(sale_price_corr)
mas_corr = sale_price_corr[(sale_price_corr.SalePrice <= -0.5) | (sale_price_corr.SalePrice >= 0.5) & (sale_price_corr.SalePrice < 1)]
mas_corr


```
